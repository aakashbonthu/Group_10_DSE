```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Gold 100years.csv")

# Convert Date column to datetime
df["Date"] = pd.to_datetime(df["Date"], dayfirst=True)

# Select one month (January 2020)
selected_month = "2020-01"

# Filter data for the selected month
month_data = df[df["Date"].dt.to_period("M") == selected_month]

# Get top 10 highest values
top10 = month_data.nlargest(10, "Value")

# Create Folder for Figures
save_folder = r"E:\UE class schedule\DATA SCIENCE\DSB Project\Figures"
os.makedirs(save_folder, exist_ok=True)

# Create labels using the day of the month
top10["Day"] = top10["Date"].dt.strftime("%d-%b")

# Plot bar chart
plt.figure(figsize=(10,6))
plt.bar(top10["Day"], top10["Value"])

plt.title("D . 24Top 10 Gold Prices - January 2020")
plt.xlabel("Day")
plt.ylabel("Gold Price")
plt.xticks(rotation=45)

plt.tight_layout()

# Save figure
save_path = os.path.join(save_folder, "D . 24Top10_Gold_January2020.png")
plt.savefig(save_path, dpi=300, bbox_inches="tight")

# Show figure
plt.show()

print(f"Chart saved successfully at:\n{save_path}")
```
