```python
# QUESTION FROM SECTION D.Currency Tracker Analytics

# Q19.Identify the top currencies by global payment share in the most recent available month

import pandas as pd
import os
import matplotlib.pyplot as plt
from IPython.display import display

# Load dataset
swift = pd.read_csv("swift_currency_tracker_all_reports.csv")

# Keep only global payment share records
payment_share = swift[
    swift["metric"].str.contains("payment", case=False, na=False)
]

# Get most recent month
latest_month = payment_share["report_month"].max()

# Filter latest month
latest_data = payment_share[
    payment_share["report_month"] == latest_month
]

# Rank currencies by payment share
ranking_table = (
    latest_data[
        ["currency_or_economy", "value", "unit"]
    ]
    .sort_values(by="value", ascending=False)
    .head(10)
    .reset_index(drop=True)
)

# Rename columns
ranking_table.columns = [
    "Currency",
    "Global Payment Share",
    "Unit"
]

# Display in Jupyter
print(f"Q19: Top Currencies by Global Payment Share ({latest_month})")
display(ranking_table)

# Create folder
save_folder = r"E:\UE class schedule\DATA SCIENCE\DSB Project\Tables"
os.makedirs(save_folder, exist_ok=True)

# Save CSV
ranking_table.to_csv(
    os.path.join(
        save_folder,
        "Q19_Top_Currencies_Payment_Share.csv"
    ),
    index=False
)

# Save PNG
fig, ax = plt.subplots(figsize=(10, 4))
ax.axis('off')

table = ax.table(
    cellText=ranking_table.values,
    colLabels=ranking_table.columns,
    cellLoc='center',
    loc='center'
)

table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.5)

plt.savefig(
    os.path.join(
        save_folder,
        "Q19_Top_Currencies_Payment_Share.png"
    ),
    bbox_inches='tight',
    dpi=300
)

plt.close()
```
