```python
# QUESTION FROM SECTION B.Descriptive Analytics  
# Q7.What were the highest and lowest Gold prices in the dataset, and in which years did they occur?

import pandas as pd
import os
import matplotlib.pyplot as plt
from IPython.display import display

# Load dataset
gold = pd.read_csv(
    r"E:\UE class schedule\DATA SCIENCE\DSB Project\Gold 100years.csv"
)

# Convert Date column to datetime
gold["Date"] = pd.to_datetime(gold["Date"], dayfirst=True)

# Highest and lowest prices
highest = gold.loc[gold["Value"].idxmax()]
lowest = gold.loc[gold["Value"].idxmin()]

# Create summary table
summary_table = pd.DataFrame({
    "Metric": ["Highest Gold Price", "Lowest Gold Price"],
    "Gold Price": [highest["Value"], lowest["Value"]],
    "Date": [
        highest["Date"].strftime("%d-%m-%Y"),
        lowest["Date"].strftime("%d-%m-%Y")
    ],
    "Year": [
        highest["Date"].year,
        lowest["Date"].year
    ]
})

# Display table in Jupyter
print("Q7: Highest and Lowest Gold Prices")
display(summary_table)

# Create folder
save_folder = r"E:\UE class schedule\DATA SCIENCE\DSB Project\Tables"
os.makedirs(save_folder, exist_ok=True)

# Save CSV
summary_table.to_csv(
    os.path.join(save_folder, "Q7_Gold_Highest_Lowest.csv"),
    index=False
)

# Convert table to PNG
fig, ax = plt.subplots(figsize=(8, 2))
ax.axis('off')

table = ax.table(
    cellText=summary_table.values,
    colLabels=summary_table.columns,
    cellLoc='center',
    loc='center'
)

table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.5)

# Save PNG
plt.savefig(
    os.path.join(save_folder, "Q7_Gold_Highest_Lowest.png"),
    bbox_inches='tight',
    dpi=300
)

plt.close()

```
