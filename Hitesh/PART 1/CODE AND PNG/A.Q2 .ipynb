```python
# QUESTION FROM SECTION A.Data Preparation and Understanding

# Q2.How many rows are available for Brent Oil, Gold, Silver and the Swift Currency Tracker data? 


import pandas as pd
import os
import matplotlib.pyplot as plt

# Load datasets
brent = pd.read_csv("Brent Oil.csv")
gold = pd.read_csv("Gold 100years.csv")
silver = pd.read_csv("silver 100 years.csv")
swift = pd.read_csv("swift_currency_tracker_all_reports.csv")

# Create summary table
data_summary = pd.DataFrame({
    "Dataset": ["Brent Oil", "Gold", "Silver", "SWIFT Currency Tracker"],
    "Number of Rows": [
        brent.shape[0],
        gold.shape[0],
        silver.shape[0],
        swift.shape[0]
    ],
    "Number of Columns": [
        brent.shape[1],
        gold.shape[1],
        silver.shape[1],
        swift.shape[1]
    ]
})

# Add total row
total_row = pd.DataFrame({
    "Dataset": ["Total"],
    "Number of Rows": [data_summary["Number of Rows"].sum()],
    "Number of Columns": ["-"]
})

data_summary = pd.concat([data_summary, total_row], ignore_index=True)

# Create folder
save_folder = r"E:\UE class schedule\DATA SCIENCE\DSB Project\Tables"
os.makedirs(save_folder, exist_ok=True)

# Save CSV
data_summary.to_csv(
    os.path.join(save_folder, "Q2_Data_Summary_Table.csv"),
    index=False
)

# Save PNG
fig, ax = plt.subplots(figsize=(8, 3))
ax.axis('off')

table = ax.table(
    cellText=data_summary.values,
    colLabels=data_summary.columns,
    cellLoc='center',
    loc='center'
)

table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.5)

plt.savefig(
    os.path.join(save_folder, "Q2_Data_Summary_Table.png"),
    bbox_inches='tight',
    dpi=300
)

plt.close()

# Display table in Jupyter
print("\nQ2 Data Summary Table:")
display(data_summary)
```
