```python
# Q22 - Timeline of Major Economic and Market Events

import pandas as pd
import os
import matplotlib.pyplot as plt
from IPython.display import display

# Create timeline table
timeline_table = pd.DataFrame({
    "Year": [
        2018,
        2019,
        2022,
        2023,
        2024,
        2025
    ],
    "Event": [
        "US-China Tariff Discussions",
        "Hormuz Strait Crisis",
        "Russia-Ukraine Sanctions",
        "Global Energy Supply Disruption",
        "BRICS Pay Discussions",
        "Central Bank Gold Accumulation"
    ],
    "Potential Impact": [
        "Trade uncertainty and commodity volatility",
        "Oil supply concerns and higher Brent prices",
        "Energy and commodity market disruption",
        "Higher energy prices and inflation pressure",
        "Reduced dependence on traditional payment systems",
        "Increased demand for gold as a reserve asset"
    ]
})

# Display table in Jupyter
print("A2: Timeline of Selected Economic Events")
display(timeline_table)

# Create folder
save_folder = r"E:\UE class schedule\DATA SCIENCE\DSB Project\Part2\Tables"
os.makedirs(save_folder, exist_ok=True)

# Save CSV
timeline_table.to_csv(
    os.path.join(save_folder, "Q22_Economic_Event_Timeline.csv"),
    index=False
)

# Save PNG
fig, ax = plt.subplots(figsize=(12, 4))
ax.axis('off')

table = ax.table(
    cellText=timeline_table.values,
    colLabels=timeline_table.columns,
    cellLoc='center',
    loc='center'
)

table.auto_set_font_size(False)
table.set_fontsize(9)
table.scale(1.2, 1.8)

plt.savefig(
    os.path.join(save_folder, "A2_Economic_Event_Timeline.png"),
    bbox_inches='tight',
    dpi=300
)

plt.show()

print("A2 timeline table saved successfully!")
```
