```python
import pandas as pd
import os
import matplotlib.pyplot as plt

# CREATE PART II FOLDERS

base_folder = r"E:\UE class schedule\DATA SCIENCE\DSB Project\Part2"

tables_folder = os.path.join(base_folder, "Tables")
figures_folder = os.path.join(base_folder, "Figures")

os.makedirs(tables_folder, exist_ok=True)
os.makedirs(figures_folder, exist_ok=True)

# Q1 - EVENT IDENTIFICATION

events = pd.DataFrame({
    "Global Event": [
        "Hormuz Crisis",
        "BRICS Pay Expansion",
        "Central Bank Gold Accumulation"
    ],
    "Potential Market Impact": [
        "Oil supply concerns can increase Brent crude prices.",
        "Alternative payment systems may reduce dependence on USD-based transactions.",
        "Higher gold demand from central banks can support gold prices."
    ],
    "Evidence Link": [
        "Related to Brent Oil price movements (Part I Commodity Analysis).",
        "Related to SWIFT currency payment trends (Part I Currency Analysis).",
        "Related to Gold price trends (Part I Gold Analysis)."
    ]
})

# Save CSV
events.to_csv(
    os.path.join(tables_folder,
                 "Q1_Event_Identification_Table.csv"),
    index=False
)

# Save PNG
fig, ax = plt.subplots(figsize=(14,4))
ax.axis('off')

tbl = ax.table(
    cellText=events.values,
    colLabels=events.columns,
    loc='center'
)

tbl.auto_set_font_size(False)
tbl.set_fontsize(9)
tbl.scale(1.2, 2)

plt.title(
    "Q1: Major Global Events Affecting Commodity and Currency Markets",
    pad=20,
    fontsize=12,
    weight='bold'
)

plt.savefig(
    os.path.join(figures_folder,
                 "Q1_Event_Identification_Table.png"),
    bbox_inches='tight',
    dpi=300
)
plt.show ()
plt.close()

print("Q1 files saved successfully.")
```
